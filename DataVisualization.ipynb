```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

# Configurations

```python
# Import data
configs = np.loadtxt('res/configurations.txt')[8:]

# Ns used
N = [7, 9, 11, 12]

# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(6, 6))

# Flatten axis
ax = ax.flatten()

for i, n in enumerate(N):
    # Get positions
    positions = configs[:n]

    # Remove from data
    configs = configs[n:]

    # Plot
    ax[i].scatter(positions[:, 0], positions[:, 1], color='red')

    # Plot circle and points
    circle = Circle((0, 0), 1, fill=False)
    ax[i].add_patch(circle)

    # Set axes limits and aspect ratio
    ax[i].set_aspect('equal')
    ax[i].set_xlim(-1.1 * 1, 1.1 * 1)
    ax[i].set_ylim(-1.1 * 1, 1.1 * 1)
    
    # Add title
    ax[i].set_title('N={}'.format(n))
    
    # Remove axis
    ax[i].set_xticks([])
    ax[i].set_yticks([])

# Give tight layout and save
fig.tight_layout()
fig.savefig('res/configurations.png', dpi=300)

```
